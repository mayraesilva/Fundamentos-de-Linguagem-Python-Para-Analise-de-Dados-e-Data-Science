## Pergunta de Negócio 1:

### Qual Cidade com Maior Valor de Venda de Produtos da Categoria 'Office Supplies'?

```python
max_office_supplies = df_dsa[df_dsa['Categoria'] == 'Office Supplies']

city_office_supplies = max_office_supplies.groupby(['Cidade'])['Valor_Venda'].apply(sum).sort_values(ascending=False)
print(city_office_supplies)
print(f'A cidade com o maior valor de venda de produtos na categoria "Office Supplies" é: {city_office_supplies.index[0]}')
```

## Pergunta de Negócio 2:

### Qual o Total de Vendas Por Data do Pedido?

Demonstre o resultado através de um gráfico de barras.

```python
dias_de_venda = df_dsa['Data_Pedido'].value_counts()
df_dsa['Data_Pedido'] = pd.to_datetime(df_dsa['Data_Pedido'], format='%d/%m/%Y')
my_formatter = DateFormatter('%d/%m/%Y')
print(dias_de_venda)
print('---------------------------------------------------------------------')
total_de_vendas_por_dia = df_dsa.groupby(['Data_Pedido'])['Valor_Venda'].sum()
print(total_de_vendas_por_dia)
print('---------------------------------------------------------------------')
print('---------------------------------------------------------------------')
plt.figure(figsize=(20, 6)) 
plt.bar(total_de_vendas_por_dia.index, total_de_vendas_por_dia.values, label='Vendas', color='teal')
plt.gca().xaxis.set_major_formatter(my_formatter)
plt.legend()
plt.title('Total de Vendas Por Data do Pedido')
plt.savefig('vendas-por-data.png', dpi=300)
plt.show()
```

## Pergunta de Negócio 3:

### Qual o Total de Vendas por Estado?

Demonstre o resultado através de um gráfico de barras.

```python
total_de_vendas_por_estado = df_dsa.groupby(['Estado'])['Valor_Venda'].sum().sort_index()

plt.figure(figsize=(20, 8))  # wider figure → more spacing between x-ticks
plt.bar(total_de_vendas_por_estado.index, total_de_vendas_por_estado.values, width=0.8, label='Vendas', color='darkorchid')
plt.legend()
plt.title('Total de Vendas Por Estado')
plt.xticks(rotation=90)
plt.savefig('vendas-por-estado.png', dpi=300)
plt.show()
```

## Pergunta de Negócio 6 (Desafio Nível Baby):

### Qual o Total de Vendas Por Segmento e Por Ano?

```python
quantos_por_seg = df_dsa['Segmento'].value_counts()
total_de_vendas_por_seg = df_dsa.groupby(['Segmento'])['Valor_Venda'].apply(sum)
df_dsa['Ano'] = df_dsa['Data_Pedido'].dt.year
print('------------------------------------------------------------')
print(quantos_por_seg)
print('------------------------------------------------------------')
print(total_de_vendas_por_seg)
print('------------------------------------------------------------')
print(df_dsa['Data_Pedido'])
print('------------------------------------------------------------')
print('------------------------------------------------------------')
print(df_dsa['Ano'].value_counts())
print('------------------------------------------------------------')

total_de_vendas_por_seg_e_ano = df_dsa.groupby(['Ano', 'Segmento'])['Valor_Venda'].apply(sum)
print(total_de_vendas_por_seg_e_ano)

print('------------------------------------------------------------')

max_seg_por_ano = total_de_vendas_por_seg_e_ano.groupby('Ano').idxmax()
print(max_seg_por_ano)
print('------------------------------------------------------------')

for tupla in max_seg_por_ano:
    print(f'Em {tupla[0]}, o segmento que mais teve vendas foi {tupla[1]} com um total de {total_de_vendas_por_seg_e_ano.loc[tupla]}')
```

## Pergunta de Negócio 7 (Desafio Nível Júnior):

Os gestores da empresa estão considerando conceder diferentes faixas de descontos e gostariam de fazer uma simulação com base na regra abaixo:

- Se o Valor_Venda for maior que 1000 recebe 15% de desconto.
- Se o Valor_Venda for menor que 1000 recebe 10% de desconto.

### Quantas Vendas Receberiam 15% de Desconto?

```python
vendas = df_dsa['Valor_Venda']
vendas_grandes = 0
vendas_pequenas = 0

for venda in vendas:
    if venda > 1000:
        vendas_grandes += 1
    else:
        vendas_pequenas += 1
        
print(f'Receberiam o desconto de 15% o total de {vendas_grandes} compras')
```

```python
df_dsa['Descontos'] = np.where(df_dsa['Valor_Venda']>1000,df_dsa['Valor_Venda'] * 0.85, df_dsa['Valor_Venda'] * 0.9 )
print(df_dsa['Valor_Venda'])
print('------------------------------------------------')
print(df_dsa['Descontos'])
print('------------------------------------------------')
desconto_maior = (df_dsa['Valor_Venda'] > 1000).value_counts()
print(desconto_maior)
print(f'Receberiam o desconto de 15% o total de {desconto_maior[1]} compras')
```

## Pergunta de Negócio 8 (Desafio Nível Master):

### Considere Que a Empresa Decida Conceder o Desconto de 15% do Item Anterior. Qual Seria a Média do Valor de Venda Antes e Depois do Desconto?

```python
media_antes_desconto = df_dsa['Valor_Venda'].mean()
print(media_antes_desconto)
print('---------------------------------------------')
df_dsa['Desconto_15'] = np.where(df_dsa['Valor_Venda']>1000,df_dsa['Valor_Venda'] * 0.85, df_dsa['Valor_Venda'])

media_pos_desconto_15 = df_dsa['Desconto_15'].mean()
print(media_pos_desconto_15)

print(f'A média antes do desconto ser concedido é de {media_antes_desconto:.2f} e após o desconto {media_pos_desconto_15:.2f}')
```
